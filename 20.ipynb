```python
# Creating a list with example temperatures for Warsaw and Miami

warsaw = [-2, 0, 5, 10, 16, 20, 22, 21, 17, 11, 5, 0]
miami = [20, 21, 22, 24, 26, 28, 29, 29, 28, 26, 23, 21]

# Creating a list with month names
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Creating a line plot
import matplotlib.pyplot as plt
import numpy as np

# Setting the figure size
plt.figure(figsize=(10,6))

# Drawing the plot for Warsaw
plt.plot(months, warsaw, label='Warsaw', color='purple', marker='o')

# Drawing the plot for Miami
plt.plot(months, miami, label='Miami', color='pink', marker='s')

# Title and axis labels
plt.title('Average Monthly Temperatures')
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')

# Y-axis range
plt.ylim(-5, 35)

# Legend
plt.legend()

# Grid
plt.grid()

# Show the plot
plt.show()

# Save as PNG
plt.savefig('temperatures.png', dpi=300)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('/content/drive/MyDrive/Data_sets/Pliki_wizualizacja/Zadania/Zadanie I - 3/Pareto Front.txt', sep='\t')
data2 = pd.read_csv('/content/drive/MyDrive/Data_sets/Pliki_wizualizacja/Zadania/Zadanie I - 3/assets.txt', sep='\t')
print(data.head())
print(data.columns)  # The higher the variance, the higher the risk; the higher the mean, the higher the return

# Sort data by mean return
data_sorted = data.sort_values(by='Mean')

# Best, middle, and worst portfolios
best = data_sorted.tail(500)
middle = data_sorted.iloc[500:-500]
worst = data_sorted.head(500)

# Create scatter plot
plt.figure(figsize=(10,6))
plt.scatter(middle['Variance'], middle['Mean'], color='lightgray', alpha=0.5, label='Others')
plt.scatter(worst['Variance'], worst['Mean'], color='green', label='500 Worst')
plt.scatter(best['Variance'], best['Mean'], color='red', label='500 Best')
plt.scatter(data2['Variance'], data2['Mean'], color='pink')
plt.xlabel('Risk')
plt.ylabel('Return')
plt.title('Relationship Between Return and Risk in Investment Portfolios')
plt.grid(True)

# Compute return and risk range to place Sharpe area label
min_mean = data['Mean'].min()
max_mean = data['Mean'].max()
min_var = data['Variance'].min()
max_var = data['Variance'].max()

# Center point of the range
center_x = (min_var + max_var) / 2
center_y = (min_mean + max_mean) / 2

# Add text labels
plt.text(data['Variance'].iloc[0], data['Mean'].iloc[0], 'High risk / high return', fontsize=9)
plt.text(data['Variance'].iloc[-1], data['Mean'].iloc[-1], 'Low risk / low return', fontsize=9)
plt.text(center_x, center_y, 'Sharpe Area', fontsize=9, ha='center')

plt.show()

# Save as PNG
plt.savefig('InvestmentPortfolios.png', dpi=300)
```

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load iris data and explore
iris = load_iris()
print(iris.keys())
print(iris.target_names)

# Create DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.columns)

# Add 'species' column
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Display DataFrame with the new column
print(df)

# Define list of attribute pairs
pairs = [
    ('sepal length (cm)', 'sepal width (cm)'),
    ('petal length (cm)', 'petal width (cm)'),
    ('sepal length (cm)', 'petal width (cm)'),
    ('petal length (cm)', 'sepal width (cm)')]

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Draw plots
# 1. Sepal length vs sepal width
for species in df['species'].unique():
    subset = df[df['species'] == species]
    axes[0, 0].scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species)
axes[0, 0].set_title('Sepal Length vs Sepal Width')
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Sepal Width (cm)')

# 2. Petal length vs petal width
for species in df['species'].unique():
    subset = df[df['species'] == species]
    axes[0, 1].scatter(subset['petal length (cm)'], subset['petal width (cm)'], label=species)
axes[0, 1].set_title('Petal Length vs Petal Width')
axes[0, 1].set_xlabel('Petal Length (cm)')
axes[0, 1].set_ylabel('Petal Width (cm)')

# 3. Sepal length vs petal width
for species in df['species'].unique():
    subset = df[df['species'] == species]
    axes[1, 0].scatter(subset['sepal length (cm)'], subset['petal width (cm)'], label=species)
axes[1, 0].set_title('Sepal Length vs Petal Width')
axes[1, 0].set_xlabel('Sepal Length (cm)')
axes[1, 0].set_ylabel('Petal Width (cm)')

# 4. Petal length vs sepal width
for species in df['species'].unique():
    subset = df[df['species'] == species]
    axes[1, 1].scatter(subset['petal length (cm)'], subset['sepal width (cm)'], label=species)
axes[1, 1].set_title('Petal Length vs Sepal Width')
axes[1, 1].set_xlabel('Petal Length (cm)')
axes[1, 1].set_ylabel('Sepal Width (cm)')

# Add legend once
axes[0, 1].legend(title='Species', loc='lower right')

# Adjust layout
plt.tight_layout()

plt.show()

# Save as PNG
plt.savefig('Iris', dpi=300)
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Alpine function
def alpine_function(x, y):
    return np.cos(x) * x * (1 + (np.cos(y) / 2))

# Define the Schubert function
def schubert_function(x, y):
    return ((np.cos(2*x)+1) + (2*np.cos(3*x+2)) + (3*np.cos(4*x+3)) +
            (4*np.cos(5*x+4)) + (5*np.cos(6*x+5)) +
            (np.cos(2*y+1)) + (2*np.cos(3*y+2)) + (3*np.cos(4*y+3)) +
            (4*np.cos(5*y+4)) + (5*np.cos(6*y+5)))

# Define the Rastrigin function
def rastrigin_function(x, y):
    return (x**2 - np.cos(18 * np.pi * x)) + (y**2 - np.cos(18 * np.pi * y))

# Define x and y ranges
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z1 = alpine_function(X, Y)
Z2 = schubert_function(X, Y)
Z3 = rastrigin_function(X, Y)

# Plot Alpine function
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, cmap='viridis', edgecolor='none')
ax.set_title('Alpine Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.savefig('Alpine_Function.png', dpi=300)

# Plot Schubert function
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z2, cmap='viridis', edgecolor='none')
ax.set_title('Schubert Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.savefig('Schubert_Function.png', dpi=300)

# Plot Rastrigin function
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z3, cmap='viridis', edgecolor='none')
ax.set_title('Rastrigin Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.savefig('Rastrigin_Function.png', dpi=300)
```
